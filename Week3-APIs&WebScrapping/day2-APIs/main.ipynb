#### Question 1: Find the questions and answers of last month.

```python
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1591488000, todate=1594123800)
from pandas.io.json import json_normalize

```

```python
flattened_questions=json_normalize(questions['items'])
flattened_questions.head()
```

```python
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers', fromdate=1591488000, todate=1594123800)
```

```python
flattened_answers=json_normalize(answers['items'])
flattened_answers.head()
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
best_question = SITE.fetch('questions', min=5, sort='votes', tagged="python", fromdate=1591488000, todate=1594123800)
best_question
```

```python
best_question1=json_normalize(best_question["items"])
best_question1.head()
```
