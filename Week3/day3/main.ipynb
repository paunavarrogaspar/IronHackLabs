`````python
secret_paths=[]

for folder in folders:
    if folder!=".gitignore":
        url = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/'+folder
        ihack_folder = requests.get(url, auth=(username, token))
        ihack_folder_json=ihack_folder.json()
        #print(folder)
        for i in ihack_folder_json:
            #print(i)
            secret_paths=secret_paths + [i['path']]
        
secret_paths
`````

`````python
joke_paths=[]
for i in secret_paths:
    path = re.findall(r"\d+/.00\d\d.scavengerhunt", i)
    joke_paths += path
    
sorted_paths=[]
for i in joke_paths:
    clean = re.sub(r"\d\d\d\d\d/.00", "", i)
    sorted_paths.append(clean)
    
sorted1= sorted(sorted_paths)
`````

`````python
joke_paths_sorted=[]
for i in sorted1:
    for e in joke_paths:
        if i in e:
            joke_paths_sorted.append(e)

joke_paths_sorted
`````

`````python
joke = []
for i in joke_paths_sorted:
    url = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/'+i
    headers = {'Accept': 'application/vnd.github.v3.raw'}
    response = requests.get(url, auth=(username, token), headers=headers)
    word = response.text
    joke.append(word)
    
joke
`````
